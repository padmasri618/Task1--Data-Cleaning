```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("Mall_Customers.csv")
print("Original Data Shape:", df.shape)

# Display missing values
print("\nMissing Values:\n", df.isnull().sum())

# Drop duplicates
df = df.drop_duplicates()
print("Shape After Removing Duplicates:", df.shape)

# Standardize text values
df['Gender'] = df['Gender'].str.strip().str.capitalize()

# Rename column headers (clean and uniform)
df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]

# Check and fix data tytypespes
print("\nData Types Before:\n", df.dtypes)
df['age'] = df['age'].astype(int)
print("\nData Types After:\n", df.dtypes)

# Save cleaned data
df.to_csv("Mall_Customers_Cleaned.csv", index=False)
print("\nCleaned dataset saved as 'Mall_Customers_Cleaned.csv'")

# Optional: Basic visualizations
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='gender')
plt.title('Gender Distribution')
plt.savefig('gender_distribution.png')
plt.show()

plt.figure(figsize=(6,4))
sns.histplot(data=df, x='age', bins=15, kde=True)
plt.title('Age Distribution')
plt.savefig('age_distribution.png')
plt.show()

plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='annual_income_(k$)', y='spending_score_(1-100)', hue='gender')
plt.title('Income vs Spending Score')
plt.savefig('income_vs_spending.png')
plt.show()
```
